```python
def compute_sum(n):
    if n <= 0:
        return 0
    s = 0
    for i in range(1, n+1):
        for ch in str(i):
            s += int(ch)
    return s

print(compute_sum(15))
print(compute_sum(5))
print(compute_sum(10))
```

```python
def sum_digits(n):
    total = 0
    
    for i in range(1, n + 1):
        while i > 0:
            total += i % 10   # نأخذ آخر رقم
            i //= 10 
            print(i//10)
            
    
    return total
 
print(sum_digits(14))   # 10
```
